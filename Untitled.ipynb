### `grid()` 函数的参数解释和代码示例

---

### 1. **b**（`bool`，可选）
- **功能**：控制网格的显示与否。设置为 `True` 时显示网格线，设置为 `False` 时隐藏网格线。
- **默认值**：`None`，默认根据当前网格状态来显示或隐藏网格。

#### 示例：

```python
import matplotlib.pyplot as plt
import numpy as np
#设置字体及符号
plt.rcParams['font.sans-serif'] = ['simhei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.grid(True)  # 显示网格线
plt.show()

# 或者关闭网格线
plt.plot(x, y)
plt.grid(False)  # 隐藏网格线
plt.show()

# 默认状态
plt.plot(x, y)
plt.grid(None)  # 自动调整
plt.show()
```
